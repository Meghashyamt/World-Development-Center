```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('cars.csv')
```

1. Make a figure with 2 by 3 subplots. Make sure that you adjusted the figure size and share x or y axes of the subplots if needed.

```python
fig=plt.figure(figsize=(8,5))
ax=fig.subplots(2,3)
ax[0,0].set_xlabel("X axis")
plt.tight_layout()
```

2. Make a bar plot with a variable you choose. Again make sure that you added x-label, y-label and title. If needed adjust the figure size again.

```python
plt.figure(figsize=(10,5))
plt.bar(data['car_model'],height=data['mpg'])
plt.xticks(rotation=90)
plt.xlabel("Car Models")
plt.ylabel("Milage")
plt.title("Car Milage");
```

3. Make at least 2 histogram plots. In one of them use different number of bins than the default values. Make sure you added title, x-label and figure sizes. Add your interpretation of the plot.

```python
plt.figure(figsize=(8,5))
plt.hist(data['mpg'],edgecolor='black',bins=8)
plt.xlabel('Mileage')
plt.ylabel('Frequency count')
plt.title("Histogram 1");
```

```python
data['mpg'].skew()   # moderately right skewed for milleage column data
# -0.5 to +0.5 fairly symentrical
# -1 to -0.5 or 0.5 to 1  moderately skewed
# -1 or +1 more then than , highly skewed
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['drat'],edgecolor='black', bins=8)
plt.xlabel('Drat')
plt.ylabel('Frequency count')
plt.title("Histogram 2");
```

```python
data['drat'].skew()  # Fairly symetrical
```

4. Pick two continuous variables and make a scatter plot. Make sure that you added title, x,y labels and legend. Add your comments and interpretations of the plot.



```python
plt.scatter(data['drat'],data['mpg'],label="drat vs mileage")
plt.xlabel('drat')
plt.ylabel('mpg')
plt.title('drat vs mileage')
plt.legend(loc=0);
# As drat is increasing the mileage of the vechile is increasing.
```
